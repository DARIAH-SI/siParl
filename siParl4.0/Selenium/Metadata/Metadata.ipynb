```python
def parse_text(second_page):
    #left-side of the split: 38. nujna seja Komisija za nadzor obveščevalnih in varnostnih služb Seja',
    ls = second_page.split("\n")[0:-1:2]
    #right-side of the split: 20. 4. 2022'
    rs = second_page.split("\n")[1:-1:2]
    
    row = []
    for l, r in list(zip(ls, rs)):
        l_orig = l
        substring1 = "srečanje"
        substring2 = "posvet"
        if substring1 in l:
            l = l.split('srečanje')
            this = str(l[0])
            that = this + 'srečanje'
            l[1] = l[1].split('Seja')[0]
            l_orig = l_orig.replace('Seja', "").replace(that, "")
            row.append([that, l_orig, 'Seja', r])
        elif substring2 in l:
            l = l.split('posvet')
            this = str(l[0])
            that = this + 'posvet'
            l[1] = l[1].split('Seja')[0]
            l_orig = l_orig.replace('Seja', "").replace(that, "")
            row.append([that, l_orig, 'Seja', r])    
        else:
            l = l.split('seja')#This then splits the "50. redna seja (skupna seja)" into l[0] = "50. redna seja", l[1] = "(skupna seja...)"
            l[0] = l[0] + 'seja'
            l[1] = l[1].split('Seja')[0]
            l_orig = l_orig.replace('Seja', "").replace(l[0], "")
            row.append([l[0], l_orig, 'Seja', r])
    
    return pd.DataFrame(row)
```
